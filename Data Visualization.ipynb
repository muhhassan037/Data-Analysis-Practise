```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=True,cmap='summer')
```

```python
top_10_countries = df.nlargest(10, 'Population')

plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Population', data=top_10_countries, palette='viridis')
plt.title('Top 10 Countries by Population')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
top_5_gdp_countries = df.nlargest(5, 'GDP per Capita')

plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='GDP per Capita', data=top_5_gdp_countries, palette='crest')
plt.title('Top 5 Countries by GDP per Capita')
plt.xlabel('Country')
plt.ylabel('GDP per Capita')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Population'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Distribution of Population Across Countries')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(y=df['GDP per Capita'].dropna(), color='lightgreen')
plt.title('Box Plot of GDP per Capita')
plt.ylabel('GDP per Capita')
plt.tight_layout()
plt.show()
```
